```python
from sklearn.metrics import precision_recall_curve
import numpy as np

# Calculate probabilities using the correct model instance
probabilities = brf.predict_proba(X_test)[:, 1]

# Calculate preceision, recall, and thresholds
precision, recall, thresholds = precision_recall_curve(y_test, probabilities)

# Calculate the cost for each threshold
costs = []
for threshold in thresholds:
    pred = (probabilities >= threshold).astype(int)
    fn_cost = np.sum((pred == 0) & (y_test == 1)) * 200  # Cost for false negatives
    fp_cost = np.sum((pred == 1) & (y_test == 0)) * 100 # Cost for false positives
    total_cost = fn_cost + fp_cost
    costs.append(total_cost)

# Find the threshold with the minimum cost
min_cost_threshold = thresholds[np.argmin(costs)]
print("Optimal threshold for minimum cost:", min_cost_threshold)
```

```python
# Correcting the model variable to rf for predicting probabilities
probabilities = brf.predict_proba(X_test)[:, 1]  # Get probabilities of the positive class
predictions = (probabilities >= min_cost_threshold).astype(int)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Accuracy:", accuracy_score(y_test, predictions))

# Print classification report to see the impact on precision, recall, and F1-score
print(classification_report(y_test, predictions))

# Print the confusion matrix
print(confusion_matrix(y_test, predictions))
```
